```python
# Import Packages
import plotly.express as px
import torch as tn
import torch as th
```

```python
# Initialize Matrix and data Dimensions
D = 100
T = 150

Encoding_Matrix = th.zeros((T,D))
EF = 10000 # Encoding Frequenze


for d in range(int(D/2)):
    for t in range(T):
        Encoding_Matrix[t,2*d] = tn.sin(tn.tensor([t/EF**(2*d/D)]))
        Encoding_Matrix[t,2*d+1] = tn.cos(tn.tensor([t/EF**((2*d)/D)]))

# Plot Postional Encoding Mask
fig = px.imshow(Encoding_Matrix,aspect="auto",color_continuous_scale="RdBu")
fig.update_layout(
    height = 500,
    width = 1000
)
fig.update_xaxes(
    title= "Embedding Dimension"
)
fig.update_yaxes(
    title = "Sequence Position"
)
fig.show()
```
