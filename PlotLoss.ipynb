```python
# Import Packages
import plotly.express as px
import torch as th
import pickle as pl
import pandas as pd
import numpy as np
```

```python
#Load loss from training the models
with open('Models/TransformerLoss.pkl', 'rb') as f:
    transformer_loss= pl.load(f)[0]
with open('Models/RnnLoss.pkl', 'rb') as f:
    rnn_loss = pl.load(f)

# Define the time
time_t = np.linspace(0,1,len(transformer_loss))
time_rnn = np.linspace(0,1,len(rnn_loss))

# Intialize loss dataframe to plot transformer loss
Transformer_df = pd.DataFrame({'Transformer Loss': transformer_loss, 'Time(H)': time_t})
Transformer_df['Transformer Loss'] = Transformer_df['Transformer Loss']/50

# Intialize loss dataframe to plot rnn loss
Rnn_df = pd.DataFrame({'RNN Loss': rnn_loss, 'Time(H)': time_rnn})
Rnn_df ['RNN Loss'] = Rnn_df['RNN Loss']/50

# Print the length of the loss
print(len(transformer_loss))
print(len(rnn_loss))
```

```python
#Plot data transformer vs rnn

# Intialize the figure
fig = px.scatter()

# Add the data
fig.add_scatter(x=Transformer_df['Time(H)'], y=Transformer_df['Transformer Loss'],name="Transformer")
fig.add_scatter(x=Rnn_df['Time(H)'],y= Rnn_df['RNN Loss'],name="RNN")

# Update the figure
fig.update_layout(legend=dict(title="Losses"))
fig.update_xaxes(
    title="Time(H)"
)
fig.update_yaxes(
    title="Training loss",type="log"
)

# Show the figure
fig.show()
```
